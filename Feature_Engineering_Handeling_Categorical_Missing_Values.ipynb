##### Ordinal Number Encoding

```python
import datetime
```

```python
todays_date=datetime.datetime.today()
```

```python
todays_date
```

```python
# Decrease the date by specified number
todays_date-datetime.timedelta(1)
```

```python
todays_date-datetime.timedelta(2)
```

```python
days_data=[todays_date-datetime.timedelta(x) for x in range(0,16)]
```

```python
days_data
```

```python
#Now this is list we need to convert it into dataframe
import pandas as pd

data=pd.DataFrame(days_data)
```

```python
data.head()
```

```python
data.columns=['Day']
```

```python
data.head()
```

```python
data['weekday']=data['Day'].dt.day_name()
data.head()
```

```python
my_day=data['weekday'].unique()
```

```python
my_day
```

```python
my_day_list=list(my_day)
```

```python
my_day_list
```

```python
ready_to_map={k:i for i,k in enumerate(my_day_list,0)}
```

```python
ready_to_map
```

```python
data['weekday']=data['weekday'].map(ready_to_map)
```

```python
data.head()
```

```python
data
```

## Count or Frequency Encoding

```python
pd.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
for feature in dataset.columns:
    print(feature,":",len(dataset[feature].unique()),"labels")
```

```python
dataset['Exterior2nd'].unique()
```

```python
dataset['Exterior2nd'].value_counts()
```

```python
ready_to_map=dataset['Exterior2nd'].value_counts().to_dict()
```

```python
ready_to_map
```

```python
dataset['Exterior2nd']=dataset['Exterior2nd'].map(ready_to_map)
```

```python
dataset['Exterior2nd'].head(10)
```

```python
dataset.head()
```

##### With the help of count/frequency method Exterior2nd column successfully converted into numeric value

## Target Guided Ordinal Encoding

```python
dataset=pd.read_csv('Titanic_train.csv',usecols=['Survived','Cabin'])
```

```python
dataset.head()
```

```python
dataset['Cabin'].unique()
```

```python
dataset['Cabin'].fillna('Missing',inplace=True)
```

```python
dataset.head()
```

```python
dataset['Cabin'].astype(str)[0]
```

```python
dataset['Cabin'].astype(str).str[0]
```

```python
dataset['Cabin']=dataset['Cabin'].astype(str).str[0]
```

```python
dataset.head()
```

```python
dataset.Cabin.unique()
```

```python
dataset.groupby('Cabin')['Survived'].mean()
```

```python
dataset.groupby('Cabin')['Survived'].mean().sort_values()
```

```python
dataset.groupby('Cabin')['Survived'].mean().sort_values().index
```

```python
ordinal_labels=dataset.groupby('Cabin')['Survived'].mean().sort_values().index
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
dataset['Cabin_ordinal_labels']=dataset['Cabin'].map(ordinal_labels2)
dataset.head()
```

## Mean Encoding

```python
mean_ordinal=dataset.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
dataset['mean_ordinal_encode']=dataset['Cabin'].map(mean_ordinal)
dataset.head()
```
